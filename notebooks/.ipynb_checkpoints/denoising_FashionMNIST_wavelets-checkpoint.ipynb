```python
# Load data

np.random.seed(31)

img_size = 32 #resize the MNIST images to this dimension 

#dataloader
from torch.utils.data import DataLoader

transform_MYULA = transforms.Compose(
    [#transforms.RandomHorizontalFlip(),
     transforms.Resize((img_size, img_size)),
     transforms.ToTensor(),
     ])

dataset = datasets.FashionMNIST(root='../data/FashionMNIST', train=False, transform=transform_MYULA, download=True)
train_loader = DataLoader(dataset, batch_size=1, shuffle=False)
print('# training minibatches = %d'%len(train_loader))

#add noise, denoise using a prox. operator
noise_sd = 0.10
# verify choice for alpha for the prox problem
alpha0 = 0.1 # 0.1

for idx, (images, labels) in enumerate(train_loader):
    x_true = images #true images
#     x_noisy = x_true + noise_sd*torch.randn(x_true.size())
    #see true, noisy, and denoised
    gt = x_true[0].detach().numpy().squeeze()
    data = gt + noise_sd*np.random.randn(*gt.shape) # x_noisy[0].detach().numpy().squeeze()
    prox = prox_l1_dwt(alpha0, data, level=2)

    plt.subplot(131)
    vmin, vmax = gt.min(), gt.max()
    plt.imshow(gt,cmap='gray',vmin=vmin, vmax=vmax)
    plt.xticks([])
    plt.yticks([])
    plt.title('true')

    plt.subplot(132)
    plt.imshow(data,cmap='gray',vmin=vmin, vmax=vmax)
    plt.xticks([])
    plt.yticks([])
    plt.title('noisy')

    plt.subplot(133)
    plt.imshow(prox,cmap='gray',vmin=vmin, vmax=vmax)
    plt.xticks([])
    plt.yticks([])
    plt.title('prox')

    plt.gcf().set_size_inches(6.0, 4.0)

    break
```

```python
#unrolled gradient network (from notebook)
n_filters = 16
tau = 0.01*torch.ones(1)

class cnn_image_space(nn.Module):
    def __init__(self, n_in_channels=2, n_out_channels = 1, n_filters=n_filters): #2 input channels, 1 for image, 1 for noisy data
        super(cnn_image_space, self).__init__()
        
        self.conv1 = nn.Conv2d(n_in_channels, out_channels=n_filters, kernel_size=3, stride=1, padding=1, bias=True)
        self.conv2 = nn.Conv2d(n_filters, n_filters, kernel_size=3, stride=1, padding=1, bias=True)
        self.conv3 = nn.Conv2d(n_filters, out_channels=1, kernel_size=3, stride=1, padding=1, bias=True)
        
        self.act1 = nn.PReLU(num_parameters=1, init=0.25)
        self.act2 = nn.PReLU(num_parameters=1, init=0.25)
        
    def forward(self, x, u):
        dx = torch.cat((x, u), dim=1)
        dx = self.act1(self.conv1(dx))
        dx = self.act2(self.conv2(dx))
        dx = self.conv3(dx)
        return dx
    
class unrolled_gradient_model(nn.Module):
    def __init__(self, niter=5, tau=tau): #niter=10 worked quite well
        super(unrolled_gradient_model, self).__init__()
        
        self.niter = niter
        self.tau = nn.Parameter(tau * torch.ones(self.niter))
        self.cnn_image_layers = nn.ModuleList([cnn_image_space() for i in range(self.niter)])
        
    def forward(self, y): #y = noisy data
        x = y
        for iteration in range(self.niter):
            x = x - self.tau[iteration] * self.cnn_image_layers[iteration](x, y) # x = x - step_size*dx
        return x

prox_l1_dwt_net = unrolled_gradient_model()
```

```python
# Make sure gradient is in correct format
data_var = noise_sd**2

# Setup problem parameters
Lf = 1/data_var
delta = 0.1 / Lf
lambd = 1/Lf # must be same as in prox learn notebook

alpha = alpha0/data_var # = alpha0*Lf


# data =100*data
def gradient(x):
#     return gradient_l2_data_fidelity(x,data)
    return gradient_l2_data_fidelity(x,data,var=data_var)

def prox_g(lambd, x):
    return prox_l1_dwt(lambd*alpha, x, level=2)

prob = Problem(shape=data.shape)

num_its = 10000
myula_mc = MYULA(problem=prob,
               num_its=num_its,
               burn_in=0,
#                thinning=20,
               x_init=data,
               gradient_f=gradient,
               prox_g=prox_g,
               delta=delta,
               lambd=lambd
               )


# Load learned proximal map
model_path = os.path.join("..","models","unrolled_gradient_pretrained_nets","l1dwt_prox_model_epoch_09.pt")
prox_l1_dwt_net.load_state_dict(torch.load(model_path))

def prox_g_net(lamb, x):
    z = torch.zeros(x_true.size())
    z[0,0,:,:] = torch.tensor(x)
    prox = prox_l1_dwt_net(z)
    return prox.detach().numpy().squeeze()

prob_net = Problem(shape=data.shape)

myula_mc_net = MYULA(problem=prob_net,
               num_its=num_its,
               burn_in=0,
               x_init=data,
               gradient_f=gradient,
               prox_g=prox_g_net,
               delta=delta,
               lambd=lambd
               )

```

```python
# Compute Markov Chain
print("MYULA with exact prox")
chain = myula_mc.compute_chain()
print("MYULA with learned prox")
chain_net = myula_mc_net.compute_chain()
```

```python
# Evaluate whether the burn-in steps and the thinning are chosen properly

import pywt

def cost(x):
    level = 2
    v = pywt.wavedec2(x, 'db4', mode='zero', level=level)  # wavelet transform x
    l1 = np.sum(np.abs(v[0]))
    for l in range(level):    
        # skip l=0
        for coeff in v[l+1]:
            l1 += np.sum(np.abs(coeff))
    
    
    return 0.5/data_var * np.sum((x - data)**2) + alpha * l1

def running_mean(chn):
    csum = np.cumsum(chn, axis=0)
    for i in range(num_its):
        csum[i] = csum[i]/(i+1)
        
    return csum
    

rmean = running_mean(chain)
rmean_net = running_mean(chain_net)
costs = []
costs_net = []
mcosts = []
mcosts_net = []

for i in range(num_its):
    costs.append(cost(chain[i]))
    costs_net.append(cost(chain_net[i]))
    mcosts.append(cost(rmean[i]))
    mcosts_net.append(cost(rmean_net[i]))
    
    
rnorm = np.sum((rmean[1:] - rmean[0:-1])**2, axis = (1,2))
rnorm_net = np.sum((rmean_net[1:] - rmean_net[0:-1])**2, axis = (1,2))
    
plt.subplot(121)
plt.plot(range(num_its-1), rnorm)
plt.plot(range(num_its-1), rnorm_net)
plt.yscale("log")
plt.ylabel("Succesive Iterate Distance")
plt.xlabel("Iterations")
plt.legend(("exact prox",
            "learned prox"))

plt.subplot(122)
plt.plot(range(num_its), costs)
plt.plot(range(num_its), costs_net)
plt.plot(range(num_its), mcosts)
plt.plot(range(num_its), mcosts_net)
plt.plot(range(num_its), [cost(prox)]*num_its)
plt.ylabel("Cost")
plt.xlabel("Iterations")
plt.legend(("Iterates with exact prox",
            "Iterates with learned prox",
            "Posterior mean with exact prox", 
            "Posterior mean with learned prox", 
            "Cost MAP solution"))


plt.gcf().set_size_inches(15.0, 6.0)
plt.savefig(os.path.join(results_dir, "denoising_FashionMNIST_wavelet_cost.eps"))
plt.show()

print("error = {}".format((cost(mean) - cost(prox))/cost(prox)))
print("error learned prox = {}".format((cost(mean_net) - cost(prox))/cost(prox)))
```

```python
# Inpainting
np.random.seed(31)

def gencoordinates(n):
    seen = set()

    x, y = np.random.randint(n), np.random.randint(n)

    while True:
        seen.add((x, y))
        yield (x, y)
        x, y = np.random.randint(n), np.random.randint(n)
        while (x, y) in seen:
            x, y = np.random.randint(n), np.random.randint(n)
            
def getmask(imsize, percentage = 0.5):
    assert percentage <= 1
    
    n = int(np.ceil(imsize**2 * percentage))
    coords = []
    g = gencoordinates(imsize)
    
    for i in range(n):
        coord = next(g)
        coords.append(coord)
        
    mask = np.ones((imsize,imsize))
    for x,y in coords:
        mask[x,y] = 0.
    return mask

mask = getmask(img_size,percentage=0.5)
```

```python
def operator(x):
    return mask * x

def adjoint(x):
    return x
```

```python
for idx, (images, labels) in enumerate(train_loader):
    x_true = images #true images
    #see true, noisy, and denoised
    gt = x_true[0].detach().numpy().squeeze()
    ip_data = mask * (gt + noise_sd*np.random.randn(*gt.shape))

    plt.subplot(121)
    vmin, vmax = gt.min(), gt.max()
    plt.imshow(gt,cmap='gray',vmin=vmin, vmax=vmax)
    plt.xticks([])
    plt.yticks([])
    plt.title('true')

    plt.subplot(122)
    plt.imshow(ip_data,cmap='gray',vmin=vmin, vmax=vmax)
    plt.xticks([])
    plt.yticks([])
    plt.title('noisy')

    plt.gcf().set_size_inches(6.0, 4.0)

    break
```

```python
# Setup problem parameters
Lf = 1/data_var
delta = 0.1 / Lf
lambd = 1/Lf # must be same as in prox learn notebook

alpha = alpha0/data_var # = alpha0*Lf

# update gradient
def ip_gradient(x):
    return gradient_l2_data_fidelity(x,ip_data,var=data_var,operator=operator,adjoint=adjoint)

ip_prob = Problem(shape=ip_data.shape)

num_its = 10000
ip_myula_mc = MYULA(problem=ip_prob,
               num_its=num_its,
               burn_in=0,
               x_init=ip_data,
               gradient_f=ip_gradient,
               prox_g=prox_g,
               delta=delta,
               lambd=lambd
               )


ip_prob_net = Problem(shape=ip_data.shape)

ip_myula_mc_net = MYULA(problem=ip_prob_net,
               num_its=num_its,
               burn_in=0,
               x_init=ip_data,
               gradient_f=ip_gradient,
               prox_g=prox_g_net,
               delta=delta,
               lambd=lambd
               )


```

```python
def ip_cost(x):
    level = 2
    v = pywt.wavedec2(x, 'db4', mode='zero', level=level)  # wavelet transform x
    l1 = np.sum(np.abs(v[0]))
    for l in range(level):    
        # skip l=0
        for coeff in v[l+1]:
            l1 += np.sum(np.abs(coeff))
    
    
    return 0.5/data_var * np.sum(mask * (x - data)**2) + alpha * l1

ip_rmean = running_mean(ip_chain)
ip_rmean_net = running_mean(ip_chain_net)
ip_costs = []
ip_costs_net = []
ip_mcosts = []
ip_mcosts_net = []

ip_rnorm = np.sum((ip_rmean[1:] - ip_rmean[0:-1])**2, axis = (1,2))
ip_rnorm_net = np.sum((ip_rmean_net[1:] - ip_rmean_net[0:-1])**2, axis = (1,2))

for i in range(num_its):
    ip_costs.append(ip_cost(ip_chain[i]))
    ip_costs_net.append(ip_cost(ip_chain_net[i]))
    ip_mcosts.append(ip_cost(ip_rmean[i]))
    ip_mcosts_net.append(ip_cost(ip_rmean_net[i]))
    
plt.subplot(121)
plt.plot(range(num_its-1), ip_rnorm)
plt.plot(range(num_its-1), ip_rnorm_net)
plt.yscale("log")
plt.ylabel("Succesive Iterate Distance")
plt.xlabel("Iterations")
plt.legend(("exact prox",
            "learned prox"))

plt.subplot(122)
plt.plot(range(num_its), ip_costs)
plt.plot(range(num_its), ip_costs_net)
plt.plot(range(num_its), ip_mcosts)
plt.plot(range(num_its), ip_mcosts_net)
plt.ylabel("Cost")
plt.xlabel("Iterations")
plt.legend(("Iterates with exact prox",
            "Iterates with learned prox",
            "Posterior mean with exact prox", 
            "Posterior mean with learned prox"))

plt.gcf().set_size_inches(15.0, 6.0)
plt.savefig(os.path.join(results_dir, "inpainting_FashionMNIST_wavelet_cost.eps"))
plt.show()
```
