```python
import numpy as np
import os

import torch
import torch.nn as nn
import torchvision
from torchvision import datasets
from torchvision import transforms
import matplotlib.pyplot as plt

from gradients.l2_data_fidelity import gradient_l2_data_fidelity
from problems import Problem
from proxes.l1_dwt import prox_l1_dwt
from solvers.myula import MYULA

import logging
logger = logging.getLogger()
logger.setLevel(logging.INFO)
```

```python
# Load data

img_size = 32 #resize the MNIST images to this dimension 

#dataloader
from torch.utils.data import DataLoader

transform_MYULA = transforms.Compose(
    [#transforms.RandomHorizontalFlip(),
     transforms.Resize((img_size, img_size)),
     transforms.ToTensor(),
     ])

dataset = datasets.FashionMNIST(root='../data/FashionMNIST', train=False, transform=transform_MYULA, download=True)
train_loader = DataLoader(dataset, batch_size=1, shuffle=False)
print('# training minibatches = %d'%len(train_loader))

#add noise, denoise using a prox. operator
noise_sd = 0.10

for idx, (images, labels) in enumerate(train_loader):
  x_true = images #true images
  x_noisy = x_true + noise_sd*torch.randn(x_true.size())
  #see true, noisy, and denoised
  gt = x_true[0].detach().numpy().squeeze()
  data = x_noisy[0].detach().numpy().squeeze()

  plt.subplot(121)
  vmin, vmax = gt.min(), gt.max()
  plt.imshow(gt,cmap='gray',vmin=vmin, vmax=vmax)
  plt.xticks([])
  plt.yticks([])
  plt.title('true')

  plt.subplot(122)
  plt.imshow(data,cmap='gray',vmin=vmin, vmax=vmax)
  plt.xticks([])
  plt.yticks([])
  plt.title('noisy')

  plt.gcf().set_size_inches(6.0, 4.0)

  break
```

```python
# Setup problem parameters
delta = 1e-4
lambd = 1. # must be same as in prox learn notebook

alpha = 0.1

# Make sure gradient is in correct format
data_var = noise_sd**2
def gradient(x):
    return gradient_l2_data_fidelity(x,data)
#     return gradient_l2_data_fidelity(x,data,var=data_var)

def prox_g(lamb, x):
    return prox_l1_dwt(lambd*alpha, x, level=2)


prob = Problem(shape=data.shape)

num_its = 10000
myula_mc = MYULA(problem=prob,
               num_its=num_its,
#                burn_in=1000,
#                thinning=20,
               x_init=data,
               gradient_f=gradient,
               prox_g=prox_g,
               delta=delta,
               lambd=lambd
               )
```

```python
# Compute Markov Chain
chain = myula_mc.compute_chain()
```

```python
# Evaluate result

prox = prox_g(alpha, data)

mean = np.mean(chain,axis=0)
var = np.var(chain,axis=0)
# print(vmax)
# print(vmin)
# print(mean)
# print(var)

plt.subplot(151)
plt.imshow(gt,cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('true')

plt.subplot(152)
plt.imshow(data,cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('noisy')

plt.subplot(153)
plt.imshow(prox,cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('prox')

plt.subplot(154)
plt.imshow(mean,cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('mean')

plt.subplot(155)
plt.imshow(var,cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('variance')

plt.gcf().set_size_inches(15.0, 4.0)
plt.show()


sample = [0,1,2,3,4]  # np.random.randint(0,high=num_its,size=5)
plt.subplot(151)
plt.imshow(chain[sample[0]],cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('sample {}'.format(sample[0]))

plt.subplot(152)
plt.imshow(chain[sample[1]],cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('sample {}'.format(sample[1]))

plt.subplot(153)
plt.imshow(chain[sample[2]],cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('sample {}'.format(sample[2]))

plt.subplot(154)
plt.imshow(chain[sample[3]],cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('sample {}'.format(sample[3]))

plt.subplot(155)
plt.imshow(chain[sample[4]],cmap='gray',vmin=vmin, vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('sample {}'.format(sample[4]))

plt.gcf().set_size_inches(15.0, 4.0)
plt.show()
```
