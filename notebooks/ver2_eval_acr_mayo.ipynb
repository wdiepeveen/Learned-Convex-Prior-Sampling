```python
## import libraries
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import os
import glob
from PIL import Image
import random

import warnings
warnings.filterwarnings(action='once')

#get the device details
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('device = %s'%device)
```

```python
#recreate the models 
n_filters = 48 #was 24 previously
num_classes = 1 #since we're solving a binary classification
filt_dim = 5
n_layers = 10

class icnn_new_model(nn.Module):
    def __init__(self, n_in_channels=1, n_filters=n_filters, filt_dim=filt_dim, n_layers=n_layers):
        super(icnn_new_model, self).__init__()
        
        self.n_layers = n_layers
        #these layers should have non-negative weights
        self.wz = nn.ModuleList([nn.Conv2d(n_filters, n_filters, kernel_size=filt_dim, stride=1, padding=2, bias=False) for i in range(self.n_layers)])
        
        #these layers can have arbitrary weights
        self.wx = nn.ModuleList([nn.Conv2d(n_in_channels, n_filters, kernel_size=filt_dim, stride=1, padding=2, bias=True) for i in range(self.n_layers+1)])
        
        #one final conv layer with nonnegative weights
        self.final_conv2d = nn.Conv2d(n_filters, 1, kernel_size=filt_dim, stride=1, padding=2, bias=False)
        
        #slope of leaky-relu
        self.negative_slope = 0.2 
        
    def forward(self, x):
        z = F.leaky_relu(self.wx[0](x), negative_slope=self.negative_slope)
        for layer in range(self.n_layers):
            z = F.leaky_relu(self.wz[layer](z) + self.wx[layer+1](x), negative_slope=self.negative_slope)
        z = self.final_conv2d(z)
        z_avg = F.avg_pool2d(z, z.size()[2:]).view(z.size()[0], -1)
        
        return z_avg
                    
#define an foe network separately
class foe_model(nn.Module):
    def __init__(self, n_in_channels=1):
        super(foe_model, self).__init__()
        #FoE kernels
        self.penalty = nn.Parameter((-12.0) * torch.ones(1).to(device))
        self.n_kernels = 10
        self.conv = nn.ModuleList([nn.Conv2d(n_in_channels, 32, kernel_size=7, stride=1, padding=3, bias=False)\
                                   for i in range(self.n_kernels)])
        
    def forward(self, x):
        #compute the output of the FoE part
        total_out = 0.0
        for kernel_idx in range(self.n_kernels):
            x_out = torch.abs(self.conv[kernel_idx](x))
            x_out_flat = x_out.view(x.size(0),-1)
            total_out += torch.sum(x_out_flat,dim=1)
        
        total_out = total_out.view(x.size(0),-1)
        return (F.softplus(self.penalty))*total_out

#define a network for training the l2 term
class l2_net_model(nn.Module):
    def __init__(self):
        super(l2_net_model, self).__init__()
        
        self.l2_penalty = nn.Parameter((-8.0) * torch.ones(1).to(device))
   
    def forward(self, x):
        l2_term = torch.sum(x.view(x.size(0), -1)**2, dim=1)
        return  (F.softplus(self.l2_penalty))*l2_term

#realize the models
acr = icnn_new_model().to(device) 
foe = foe_model().to(device)
l2_net = l2_net_model().to(device) 

#load the models
pretrained_network_path = './ver2_pretrained_nets/'


# #when no FoE is used
# acr.load_state_dict(torch.load(pretrained_network_path + "acr_new_model.pt"))
# #foe.load_state_dict(torch.load(pretrained_network_path + "foe_net.pt"))
# l2_net.load_state_dict(torch.load(pretrained_network_path + "l2_term.pt"))

#when FoE is used
# acr.load_state_dict(torch.load(pretrained_network_path + "acr_new_model_foe_included_epoch_09.pt"))
# foe.load_state_dict(torch.load(pretrained_network_path + "foe_net_foe_included_epoch_09.pt"))
# l2_net.load_state_dict(torch.load(pretrained_network_path + "l2_term_foe_included_epoch_09.pt"))

acr.load_state_dict(torch.load(pretrained_network_path + "acr_new_model_foe_included_paired.pt"))
foe.load_state_dict(torch.load(pretrained_network_path + "foe_net_foe_included_paired.pt"))
l2_net.load_state_dict(torch.load(pretrained_network_path + "l2_term_foe_included_paired.pt"))


print('total number of parameters in ACR = %d'%sum(p.numel() for p in acr.parameters()))
print('l2_factor = %.4e'%((F.softplus(l2_net.l2_penalty))).item()) 
```

```python
#eval data
from torchvision import transforms
from torch.utils.data import DataLoader
from torch.utils.data import Dataset

#custom dataset class 
class ImageDataset(Dataset):
    def __init__(self, root, transforms_= None, unaligned = True, mode = 'train'):
        self.transform = transforms.Compose(transforms_)
        self.unaligned = unaligned

        self.files_A = sorted(glob.glob(os.path.join(root, '%s/filtered_back_projection'% mode) + '/*.*'))
        self.files_B = sorted(glob.glob(os.path.join(root, '%s/groundtruth'% mode) + '/*.*'))


    def __getitem__(self, index):
        fbp = self.transform(Image.fromarray(np.load(self.files_A[index % len(self.files_A)])))
        
        if self.unaligned:
            ground_truth = self.transform(Image.fromarray(np.load(self.files_B[random.randint(0, len(self.files_B) - 1)])))
        else:
            ground_truth = self.transform(Image.fromarray(np.load(self.files_B[index % len(self.files_B)])))
        
        

        return {'fbp': fbp, 'ground_truth': ground_truth}

    def __len__(self):
        return max(len(self.files_A), len(self.files_B))

val_batch_size = 1
transform_train_and_test = [transforms.ToTensor()]
datapath = './mayo_data_arranged_patientwise/'
val_loader = DataLoader(ImageDataset(datapath, transforms_=transform_train_and_test,\
                                       unaligned=False, mode = 'test'), batch_size = val_batch_size, shuffle = False) 
print('no. of test samples = %d'%len(val_loader))
```

```python
#ray transform construction
import odl
from odl import Operator
#ray transform construction
class OperatorFunction(torch.autograd.Function):

    """Wrapper of an ODL operator as a ``torch.autograd.Function``.
    """

    @staticmethod
    def forward(ctx, operator, input):
        """Evaluate forward pass on the input.
        Parameters
        ----------
        ctx : context object
            Object to communicate information between forward and backward
            passes.
        operator : `Operator`
            ODL operator to be wrapped. For gradient computations to
            work, ``operator.derivative(x).adjoint`` must be implemented.
        input : `torch.Tensor`
            Point at which to evaluate the operator.
        Returns
        -------
        result : `torch.Tensor`
            Tensor holding the result of the evaluation.
        """
        if not isinstance(operator, Operator):
            raise TypeError(
                "`operator` must be an `Operator` instance, got {!r}"
                "".format(operator)
            )

        # Save operator for backward; input only needs to be saved if
        # the operator is nonlinear (for `operator.derivative(input)`)
        ctx.operator = operator

        if not operator.is_linear:
            # Only needed for nonlinear operators
            ctx.save_for_backward(input)

        # TODO(kohr-h): use GPU memory directly when possible
        # TODO(kohr-h): remove `copy_if_zero_strides` when NumPy 1.16.0
        # is required
        input_arr = copy_if_zero_strides(input.cpu().detach().numpy())

        # Determine how to loop over extra shape "left" of the operator
        # domain shape
        in_shape = input_arr.shape
        op_in_shape = operator.domain.shape
        if operator.is_functional:
            op_out_shape = ()
            op_out_dtype = operator.domain.dtype
        else:
            op_out_shape = operator.range.shape
            op_out_dtype = operator.range.dtype

        extra_shape = in_shape[:-len(op_in_shape)]
        if in_shape[-len(op_in_shape):] != op_in_shape:
            shp_str = str(op_in_shape).strip('(,)')
            raise ValueError(
                'input tensor has wrong shape: expected (*, {}), got {}'
                ''.format(shp_str, in_shape)
            )

        # Store some information on the context object
        ctx.op_in_shape = op_in_shape
        ctx.op_out_shape = op_out_shape
        ctx.extra_shape = extra_shape
        ctx.op_in_dtype = operator.domain.dtype
        ctx.op_out_dtype = op_out_dtype

        # Evaluate the operator on all inputs in a loop
        if extra_shape:
            # Multiple inputs: flatten extra axes, then do one entry at a time
            input_arr_flat_extra = input_arr.reshape((-1,) + op_in_shape)
            results = []
            for inp in input_arr_flat_extra:
                results.append(operator(inp))

            # Stack results, reshape to the expected output shape and enforce
            # correct dtype
            result_arr = np.stack(results).astype(op_out_dtype, copy=False)
            result_arr = result_arr.reshape(extra_shape + op_out_shape)
        else:
            # Single input: evaluate directly
            result_arr = np.asarray(
                operator(input_arr)
            ).astype(op_out_dtype, copy=False)

        # Convert back to tensor
        tensor = torch.from_numpy(result_arr).to(input.device)
        return tensor

    @staticmethod
    def backward(ctx, grad_output):
        
        # Return early if there's nothing to do
        if not ctx.needs_input_grad[1]:
            return None, None

        operator = ctx.operator

        # Get `operator` and `input` from the context object (the input
        # is only needed for nonlinear operators)
        if not operator.is_linear:
            # TODO: implement directly for GPU data
            # TODO(kohr-h): remove `copy_if_zero_strides` when NumPy 1.16.0
            # is required
            input_arr = copy_if_zero_strides(
                ctx.saved_tensors[0].detach().cpu().numpy()
            )

        # ODL weights spaces, pytorch doesn't, so we need to handle this
        try:
            dom_weight = operator.domain.weighting.const
        except AttributeError:
            dom_weight = 1.0
        try:
            ran_weight = operator.range.weighting.const
        except AttributeError:
            ran_weight = 1.0
        scaling = dom_weight / ran_weight

        # Convert `grad_output` to NumPy array
        grad_output_arr = copy_if_zero_strides(
            grad_output.detach().cpu().numpy()
        )

        # Get shape information from the context object
        op_in_shape = ctx.op_in_shape
        op_out_shape = ctx.op_out_shape
        extra_shape = ctx.extra_shape
        op_in_dtype = ctx.op_in_dtype

        # Check if `grad_output` is consistent with `extra_shape` and
        # `op_out_shape`
        if grad_output_arr.shape != extra_shape + op_out_shape:
            raise ValueError(
                'expected tensor of shape {}, got shape {}'
                ''.format(extra_shape + op_out_shape, grad_output_arr.shape)
            )

        # Evaluate the (derivative) adjoint on all inputs in a loop
        if extra_shape:
            # Multiple gradients: flatten extra axes, then do one entry
            # at a time
            grad_output_arr_flat_extra = grad_output_arr.reshape(
                (-1,) + op_out_shape
            )

            results = []
            if operator.is_linear:
                for ograd in grad_output_arr_flat_extra:
                    results.append(np.asarray(operator.adjoint(ograd)))
            else:
                # Need inputs, flattened in the same way as the gradients
                input_arr_flat_extra = input_arr.reshape((-1,) + op_in_shape)
                for ograd, inp in zip(
                    grad_output_arr_flat_extra, input_arr_flat_extra
                ):
                    results.append(
                        np.asarray(operator.derivative(inp).adjoint(ograd))
                    )

            # Stack results, reshape to the expected output shape and enforce
            # correct dtype
            result_arr = np.stack(results).astype(op_in_dtype, copy=False)
            result_arr = result_arr.reshape(extra_shape + op_in_shape)
        else:
            # Single gradient: evaluate directly
            if operator.is_linear:
                result_arr = np.asarray(
                    operator.adjoint(grad_output_arr)
                ).astype(op_in_dtype, copy=False)
            else:
                result_arr = np.asarray(
                    operator.derivative(input_arr).adjoint(grad_output_arr)
                ).astype(op_in_dtype, copy=False)

        # Apply scaling, convert to tensor and return
        if scaling != 1.0:
            result_arr *= scaling
        grad_input = torch.from_numpy(result_arr).to(grad_output.device)
        return None, grad_input  # return `None` for the `operator` part


class OperatorModule(torch.nn.Module):

    """Wrapper of an ODL operator as a ``torch.nn.Module``.
    """

    def __init__(self, operator):
        """Initialize a new instance."""
        super(OperatorModule, self).__init__()
        self.operator = operator

    def forward(self, x):
        """Compute forward-pass of this module on ``x``.
        """
        in_shape = tuple(x.shape)
        in_ndim = len(in_shape)
        op_in_shape = self.operator.domain.shape
        op_in_ndim = len(op_in_shape)
        if in_ndim <= op_in_ndim or in_shape[-op_in_ndim:] != op_in_shape:
            shp_str = str(op_in_shape).strip('()')
            raise ValueError(
                'input tensor has wrong shape: expected (N, *, {}), got {}'
                ''.format(shp_str, in_shape)
            )
        return OperatorFunction.apply(self.operator, x)

    def __repr__(self):
        """Return ``repr(self)``."""
        op_name = self.operator.__class__.__name__
        op_in_shape = self.operator.domain.shape
        if len(op_in_shape) == 1:
            op_in_shape = op_in_shape[0]
        op_out_shape = self.operator.range.shape
        if len(op_out_shape) == 1:
            op_out_shape = op_out_shape[0]

        return '{}({}) ({} -> {})'.format(
            self.__class__.__name__, op_name, op_in_shape, op_out_shape
        )


def copy_if_zero_strides(arr):
    """Workaround for NumPy issue #9165 with 0 in arr.strides."""
    assert isinstance(arr, np.ndarray)
    return arr.copy() if 0 in arr.strides else arr


#compute the CT forward operator
import odl
num_angles = 200 #previously 50 
num_detector_pixels = 400 #previously 200 
noise_sd = 2.0

img_size = 512
space = odl.uniform_discr([-128, -128], [128, 128], [img_size, img_size], dtype='float32', weighting=1.0)
angle_partition = odl.uniform_partition(0, 2 * np.pi, num_angles)
detector_partition = odl.uniform_partition(-360, 360, num_detector_pixels)
#geometry = odl.tomo.FanFlatGeometry(angle_partition, detector_partition, src_radius=500, det_radius=500)
geometry = odl.tomo.parallel_beam_geometry(space, num_angles=num_angles, det_shape=num_detector_pixels)
#geometry = odl.tomo.geometry.conebeam.cone_beam_geometry(space, src_radius=500, det_radius=500, num_angles=num_angles, det_shape=num_detector_pixels)


ray_transform_operator = odl.tomo.RayTransform(space, geometry)
#fbp_operator = odl.tomo.fbp_op(ray_transform_operator, filter_type='Hann', frequency_scaling=0.9)
fbp_operator = odl.tomo.fbp_op(ray_transform_operator)
ray_transform_adjoint = ray_transform_operator.adjoint

# Wrap ODL operator as `Module`
op = OperatorModule(ray_transform_operator).to(device)
fbp_op = OperatorModule(fbp_operator).to(device)
op_adj = OperatorModule(ray_transform_adjoint).to(device)

print('num_of_angles=%d'%geometry.angles.size)
print('num_of_detector_pixels=%d'%geometry.detector.size)

op_norm = 1.1 * odl.power_method_opnorm(ray_transform_operator)
print('op_norm = %.4f'%op_norm)
```

```python
#some useful routines
from skimage.measure import compare_ssim, compare_psnr
sq_loss = torch.nn.MSELoss(reduction='mean')

#hard clip image to a specific interval
def cut_image(im, vmin, vmax):
    im = np.maximum(im, vmin)
    im = np.minimum(im, vmax)
    return im

#convex reg optimizer
def acr_optimizer(x_init, x_ground_truth, y_test, n_iter, lambda_acr, lr=0.8): #previously lr=0.01
    x_cvx = x_init.clone().detach().requires_grad_(True).to(device) 
    x_optimizer = torch.optim.SGD([x_cvx], lr=lr)
    lr_scheduler = torch.optim.lr_scheduler.StepLR(x_optimizer, step_size=20, gamma=0.95)
    x_test_np = x_ground_truth.cpu().detach().numpy()
    data_range = np.max(x_test_np) - np.min(x_test_np)
    
    for iteration in np.arange(n_iter):
        x_optimizer.zero_grad()
        y_cvx = op(x_cvx)
        data_loss = sq_loss(y_test, y_cvx)
        
#         #when no foe is used
#         prior =  lambda_acr * (acr(x_cvx).mean() + l2_net(x_cvx).mean())
        
        #when foe is used
        prior =  lambda_acr * (acr(x_cvx).mean() + foe(x_cvx).mean() + l2_net(x_cvx).mean())
        
        #check individual contributions
        prior_acr = lambda_acr*acr(x_cvx).mean()
        prior_foe = lambda_acr*foe(x_cvx).mean()
        prior_l2 = lambda_acr*l2_net(x_cvx).mean()
        
        variational_loss = data_loss + prior

        variational_loss.backward(retain_graph=True)
        x_optimizer.step()
        #lr_scheduler.step()

        x_np = x_cvx.cpu().detach().numpy().squeeze()
        psnr = compare_psnr(np.squeeze(x_test_np),x_np,data_range=data_range)
        ssim = compare_ssim(np.squeeze(x_test_np),x_np,data_range=data_range)
        
        if(iteration%50==0):
            print('PSNR-convex = %.4f'%psnr)
            print('variational loss = {:.6f}, data term = {:.6f}, prior = {:.6f}, prior_acr = {:.6f}, prior_foe = {:.6f}, prior_l2 = {:.6f}'\
                  .format(variational_loss.item(),data_loss.item(), prior.item(), prior_acr.item(), prior_foe.item(), prior_l2.item()))
            print('-----------------')
    
    x_np = x_cvx.cpu().detach().numpy().squeeze()
    x_np = cut_image(x_np, vmin=0.0, vmax=1.0)
    psnr = compare_psnr(np.squeeze(x_test_np),x_np,data_range=data_range)
    ssim = compare_ssim(np.squeeze(x_test_np),x_np,data_range=data_range)
    return x_np, psnr, ssim
```

```python
#display setting
import matplotlib.pyplot as plt
font_size = 16
cmap_name = 'bone'

noise_sd = 2.0

#hyper-params in the variational problem
lambda_acr = 0.05 #optimal value yet to be found! (previously used: 0.05)
n_iter = 400

#save the output images
def plot_and_save_image(im, idx, path, psnr, ssim, recon_method = 'acr', cmap_name=cmap_name):
    plt.imshow(im,cmap=cmap_name,vmin=0.0,vmax=1.0) #windowing: vmin=0.0,vmax=0.50
    plt.gcf().set_size_inches(5.0,5.0)
    
    plt.xticks([])
    plt.yticks([])
    os.makedirs(path, exist_ok=True)
    plt.savefig(path + recon_method + '_' + '%03d'%(idx+1) + '.png', bbox_inches='tight', transparent = False, pad_inches=0.1)
    plt.show()

#compute the avg PSNR and SSIM too
psnr_avg_acr = 0.0
ssim_avg_acr = 0.0
scan_count = 0

for idx, batch in enumerate(val_loader):
    if(idx%10==0): #save every 10th image (idx%10==0)
        print('######### test sample = %d #########'%(idx+1))
        img = batch["ground_truth"].to(device) #true images
        x_test = img.view(1,1,img_size,-1)
        x_test_np = x_test.cpu().detach().numpy()
        x_test_np_squeezed = x_test_np.squeeze()
        data_range = np.max(x_test_np) - np.min(x_test_np)
        
        
    
        #get the FBP reconstruction
        clean_data = op(x_test)
        rand_noise = noise_sd*torch.randn(clean_data.size()).to(device)
        snr_data = 10*torch.log10(torch.sum(clean_data**2)/torch.sum(rand_noise**2))
        snr_data_sigma = 10*torch.log10(torch.mean(clean_data**2)/(noise_sd**2))
        print('SNR in the data space = %.4f dB'%snr_data_sigma)
        y_test = clean_data + rand_noise
        x_fbp = fbp_op(y_test)
        x_fbp_np = x_fbp.cpu().detach().numpy().squeeze()
        x_fbp_np = cut_image(x_fbp_np,vmin=0.0,vmax=1.0)
        

        #convex reg. reconstruction and FBP
        x_init = torch.from_numpy(x_fbp_np).view(x_fbp.size()).to(device)
        x_np_cvx, psnr_cvx, ssim_cvx = acr_optimizer(x_init, x_test, y_test, n_iter, lambda_acr)
        psnr_fbp = compare_psnr(x_test_np_squeezed,x_fbp_np,data_range=data_range)
        ssim_fbp = compare_ssim(x_test_np_squeezed,x_fbp_np,data_range=data_range)
        print("PSNR of FBP = {:.4f} dB, SSIM of FBP = {:.4f}".format(psnr_fbp, ssim_fbp))
        print("PSNR of ACR (paired) = {:.4f} dB, SSIM of ACR (paired) = {:.4f}".format(psnr_cvx, ssim_cvx))

        #save ACR reconstruction
        acr_path = './coupled_AR_training_images/acr_foe_included_paired/'
        os.makedirs(acr_path, exist_ok=True)
        plot_and_save_image(x_np_cvx, idx, acr_path, psnr_cvx, ssim_cvx, 'acr_foe_included_paired', cmap_name=cmap_name)

        #update the avg 
        psnr_avg_acr += psnr_cvx
        ssim_avg_acr += ssim_cvx
        scan_count += 1
```
